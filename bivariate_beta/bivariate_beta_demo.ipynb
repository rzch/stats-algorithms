# Bivarite Beta Demo

This notebook demonstrates sampling and estimation of a bivariate beta distribution, as described in [Arnold & Ng, 2011 - Flexible bivariate beta distributions](https://10.1016/j.jmva.2011.04.001).

The distribution is parametrised by 5 parameters, $\left(\alpha_{1}, \dots, \alpha_{5}\right)$.

```python
import matplotlib.pyplot as plt
import numpy as np
import bivariate_beta
```

We sample some data from the distribution using some nominal parameter values.

```python
n = 10000

alpha = np.array([8, 8, 5, 5, 3])

xsample, ysample = bivariate_beta.sample(n, alpha[0], alpha[1], alpha[2], alpha[3], alpha[4])

plt.hist2d(xsample, ysample, bins = 40)
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.show()
```

## Estimation of Parameters by Method of Moments

We implement the method of moments estimator as described in Section 3.2. The estimator recovers roughly the same parameters, and we can visualise the estimated density.

```python
alpha1, alpha2, alpha3, alpha4, alpha5 = bivariate_beta.moment_estimator(xsample, ysample)
print((alpha1, alpha2, alpha3, alpha4, alpha5))

n = 100000

xsample, ysample = bivariate_beta.sample(n, alpha1, alpha2, alpha3, alpha4, alpha5)

plt.hist2d(xsample, ysample, bins = 40)
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.show()
```

## Estimation of Parameters by Modified Maximum Likelihood

We implement the modified maximum likelihood estimator as described in Section 3.1. The estitator recovers roughly the same parameters (slightly closer to the true values compared to the method of moments estimator), and we can visualise the esimated density.

```python
alpha1, alpha2, alpha3, alpha4, alpha5 = bivariate_beta.modified_ml_estimator(xsample, ysample)
print((alpha1, alpha2, alpha3, alpha4, alpha5))

n = 100000

xsample, ysample = bivariate_beta.sample(n, alpha1, alpha2, alpha3, alpha4, alpha5)

plt.hist2d(xsample, ysample, bins = 40)
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.show()
```
